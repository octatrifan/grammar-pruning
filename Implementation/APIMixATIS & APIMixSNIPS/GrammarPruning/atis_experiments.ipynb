```python
import re
import json
param_values = {}
for func_params in schema.values():
    for param, vals in func_params.items():
        param_values.setdefault(param, set()).update(vals)

def extract_input_items(text):
    """
    Scan the input text once per parameter, so duplicates only
    appear if the user actually wrote them multiple times.
    """
    found = {}
    lc = text.lower()
    for param, vals in param_values.items():
        # match longest strings first
        for v in sorted(vals, key=len, reverse=True):
            # find all occurrences
            for _ in re.finditer(r'\b' + re.escape(v.lower()) + r'\b', lc):
                found.setdefault(param, []).append(v)
    return found

def extract_output_items(calls):
    """
    Parse each gold function-call string into a dict of its parameters.
    """
    results = []
    pat = re.compile(r"(\w+)\(([^)]*)\)")
    for call in calls:
        m = pat.match(call)
        if not m:
            continue
        func, body = m.groups()
        params = {}
        for part in re.split(r",\s*", body):
            if "=" in part:
                k, v = part.split("=", 1)
                params[k.strip()] = v.strip().strip("'")
        results.append({func: params})
    return results

def augment_data(in_path="test_manual.json", out_path="atis_data_augmented.json"):
    # load the whole list at once
    with open(in_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    with open(out_path,'w',encoding='utf-8') as out:
        for entry in data:
            # skip anything that isn’t already an object
            if not isinstance(entry, dict):
                continue
            entry["input_items"]  = extract_input_items(entry.get("input", ""))
            entry["output_items"] = extract_output_items(entry.get("output", []))
            out.write(json.dumps(entry) + "\n")
    print(f"Augmented data written to {out_path}")

# Run immediately
if __name__ == "__main__":
    augment_data()
```

```python
#SIMPLE CONSTRAINED DECODING GRAMMAR

import guidance
from guidance import select, capture


for intent, slots in schema.items():
    for slot, values in slots.items():
        list_name = f"valid_{intent}_{slot}"
        globals()[list_name] = [f"'{val}'" for val in values] + ["None"]

for intent, slots in schema.items():
    for slot in slots:
        list_name = f"valid_{intent}_{slot}"
        print(f"{list_name} = {globals()[list_name]}")

def consume_once(param, val):
    pass


# === all “valid” lists ===
valid_atis_airport_city              = ["'ontario'","'new york city'","'la'","None"]
valid_atis_airfare_departure_city    = ["'pittsburgh'","'nashville'","'la'","None"]
valid_atis_airfare_destination_city  = ["'la'","'seattle'","'montreal'","None"]
valid_atis_city_city                 = ["'la'","'denver'","None"]

# atis_airport
valid_atis_airport_state             = ["'california'", "'arizona'", "None"]
valid_atis_airport_modifier          = ["'closest'", "None"]
valid_atis_airport_city              = ["'ontario'", "'new york city'", "'la'", "None"]

# atis_city
valid_atis_city_city                 = ["'la'", "None"]
valid_atis_city_airline              = ["'spirit'", "'eastwest'", "'northeast'", "'northwest'", "None"]

# atis_quantity
valid_atis_quantity_airline          = [
    "'canadian airlines international'",
    "'canadian airlines'",
    "'british airways'",
    "'us air canadian airlines international'",
    "None"
]
valid_atis_quantity_aircraft         = ["'dh8'", "'ca31'", "'320'", "'j31'", "None"]

# atis_airfare
valid_atis_airfare_departure_city    = [
    "'pittsburgh'", "'nashville'", "'boston'", "'columbus'",
    "'toronto'", "'detroit'", "'indianapolis'", "'washington'",
    "None"
]
valid_atis_airfare_destination_city  = [
    "'atlanta'", "'cleveland'", "'st. louis'", "'san diego'",
    "'seattle'", "'memphis'", "'washington'", "'st. petersburg'", "'montreal'",
    "None"
]
valid_atis_airfare_airline_code      = ["'twa'", "None"]
valid_atis_airfare_class             = ["'first class'", "'coach'", "None"]
valid_atis_airfare_relative_cost     = ["'under'", "None"]
valid_atis_airfare_fare              = ["'200 dollars'", "None"]
valid_atis_airfare_departure_day     = ["'saturday'", "None"]

# atis_flight_no
valid_atis_flight_no_departure_city      = ["'phoenix'", "'cleveland'", "'oakland'", "'dallas'", "'houston'", "'chicago'", "None"]
valid_atis_flight_no_destination_city    = ["'seattle'", "'dallas'", "'salt lake city'", "'milwaukee'", "'houston'", "None"]
valid_atis_flight_no_airline             = ["'continental'", "'american airlines'", "None"]
valid_atis_flight_no_relative_departure_time = ["'before'", "None"]
valid_atis_flight_no_departure_time      = ["'8 am'", "None"]
valid_atis_flight_no_departure_day       = ["'thursday'", "None"]

# atis_capacity
valid_atis_capacity_airline          = ["'delta'", "'northwest'", "None"]
valid_atis_capacity_departure_city   = ["'san francisco'", "'seattle'", "None"]
valid_atis_capacity_destination_city = ["'salt lake city'", "'houston'", "None"]
valid_atis_capacity_aircraft         = ["'be1'", "'m80'", "'l1011'", "'73s'", "'757'", "'733'", "None"]

# atis_distance
valid_atis_distance_departure_city   = ["'san francisco'", "'los angeles'", "'dallas'", "'boston'", "'toronto'", "'baltimore'", "None"]
valid_atis_distance_destination_city = ["'new york'", "'san francisco'", "'los angeles'", "'dallas'", "'toronto'", "'ottawa'", "None"]

# atis_ground_service
valid_atis_ground_service_transport_type = ["'car'", "'limousine'", "None"]
valid_atis_ground_service_city            = [
    "'new york city'", "'fort worth'", "'tampa'", "'st. louis'",
    "'denver'", "'los angeles'", "'dallas'", "'salt lake city'",
    "'milwaukee'", "'washington'", "'las vegas'", "None"
]

# atis_abbreviation
valid_atis_abbreviation_fare_code    = ["'f'", "'m'", "'qo'", "'q'", "None"]
valid_atis_abbreviation_airline_code = ["'ua'", "None"]
valid_atis_abbreviation_aircraft     = ["'d9s'", "None"]

# atis_meal
valid_atis_meal_flight_number        = ["'AA665'", "'BA665'", "'LH468'", "'AA811'", "'TK382'", "None"]
valid_atis_meal_meal_type            = ["'meals'", "None"]
valid_atis_meal_departure_day        = ["'tuesday'", "None"]
valid_atis_meal_departure_period     = ["'morning'", "None"]
valid_atis_meal_meal_description     = ["'snacks'", "None"]
valid_atis_meal_airline              = ["'tower air'", "None"]

# atis_flight
valid_atis_flight_departure_city      = ["'pittsburgh'", "'memphis'", "'cleveland'", "'montreal'", "None"]
valid_atis_flight_destination_city    = ["'cleveland'", "'burbank'", "'indianapolis'", "'milwaukee'", "'long beach'", "None"]
valid_atis_flight_departure_month     = ["'april'", "None"]
valid_atis_flight_departure_day_number= ["'fifth'", "None"]
valid_atis_flight_airline             = ["'alaska airlines'", "None"]

# atis_flight_time
valid_atis_flight_time_departure_city   = ["'michigan'", "'detroit'", "None"]
valid_atis_flight_time_destination_city = ["'michigan'", "'san francisco'", "'westchester'", "'new york'", "None"]
valid_atis_flight_time_flight_time      = ["'arrival'", "'departure'", "None"]

# atis_ground_fare
valid_atis_ground_fare_transport_type = ["'limousine'", "None"]
valid_atis_ground_fare_city           = ["'michigan'", "'denver'", "'san diego'", "'los angeles'", "'boston'", "'bucharest'", "None"]

# atis_airline
valid_atis_airline_airline_code       = ["'AA'", "'as'", "'hp'", "'BA'", "'us'", "'aa'", "None"]
valid_atis_airline_departure_city     = ["'seattle'", "'toronto'", "'washington'", "None"]
valid_atis_airline_destination_city   = ["'salt lake city'", "'columbus'", "'san diego'", "'st. louis'", "None"]

# atis_day_name
valid_atis_day_name_departure_city = ["'nashville'", "'san jose'", "None"]
valid_atis_day_name_destination_city = ["'nashville'", "'tacoma'", "None"]

# atis_aircraft
valid_atis_aircraft_aircraft               = ["'m80'", "'m81'", "None"]
valid_atis_aircraft_departure_city         = ["'cleveland'", "None"]
valid_atis_aircraft_destination_city       = ["'dallas'", "None"]
valid_atis_aircraft_relative_departure_time = ["'before'", "None"]
valid_atis_aircraft_departure_period       = ["'noon'", "None"]
valid_atis_aircraft_airline                = ["'american northwest'", "None"]
valid_atis_aircraft_departure_time         = ["'4:19pm'", "None"]



@guidance(stateless=False)
def atis_airport(lm):
    first = True
    if valid_atis_airport_state != ["None"]:
        lm += "state=" + capture(select(valid_atis_airport_state), name='stateName')
        consume_once("state", lm['stateName'])
        first = False
    if valid_atis_airport_modifier != ["None"]:
        if not first: lm += ", "
        lm += "modifier=" + capture(select(valid_atis_airport_modifier), name='modifierName')
        consume_once("modifier", lm['modifierName'])
        first = False
    if valid_atis_airport_city != ["None"]:
        if not first: lm += ", "
        lm += "city=" + capture(select(valid_atis_airport_city), name='cityName')
        consume_once("city", lm['cityName'])
    return lm + ")"

@guidance(stateless=False)
def atis_city(lm):
    first = True
    if valid_atis_city_city != ["None"]:
        lm += "city=" + capture(select(valid_atis_city_city), name='cityName')
        consume_once("city", lm['cityName'])
        first = False
    if valid_atis_city_airline != ["None"]:
        if not first: lm += ", "
        lm += "airline=" + capture(select(valid_atis_city_airline), name='cityAirlineName')
        consume_once("airline", lm['cityAirlineName'])
    return lm + ")"

@guidance(stateless=False)
def atis_quantity(lm):
    first = True
    if valid_atis_quantity_airline != ["None"]:
        lm += "airline=" + capture(select(valid_atis_quantity_airline), name='qAirlineName')
        consume_once("airline", lm['qAirlineName'])
        first = False
    if valid_atis_quantity_aircraft != ["None"]:
        if not first: lm += ", "
        lm += "aircraft=" + capture(select(valid_atis_quantity_aircraft), name='qAircraftName')
        consume_once("aircraft", lm['qAircraftName'])
    return lm + ")"

@guidance(stateless=False)
def atis_airfare(lm):
    first = True
    if valid_atis_airfare_departure_city != ["None"]:
        lm += "departure_city=" + capture(select(valid_atis_airfare_departure_city), name='depCityName')
        consume_once("departure_city", lm['depCityName'])
        first = False
    if valid_atis_airfare_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_airfare_destination_city), name='destCityName')
        consume_once("destination_city", lm['destCityName'])
        first = False
    if valid_atis_airfare_airline_code != ["None"]:
        if not first: lm += ", "
        lm += "airline_code=" + capture(select(valid_atis_airfare_airline_code), name='airlineCodeName')
        consume_once("airline_code", lm['airlineCodeName'])
        first = False
    if valid_atis_airfare_class != ["None"]:
        if not first: lm += ", "
        lm += "class=" + capture(select(valid_atis_airfare_class), name='className')
        consume_once("class", lm['className'])
        first = False
    if valid_atis_airfare_relative_cost != ["None"]:
        if not first: lm += ", "
        lm += "relative_cost=" + capture(select(valid_atis_airfare_relative_cost), name='relCostName')
        consume_once("relative_cost", lm['relCostName'])
        first = False
    if valid_atis_airfare_fare != ["None"]:
        if not first: lm += ", "
        lm += "fare=" + capture(select(valid_atis_airfare_fare), name='fareName')
        consume_once("fare", lm['fareName'])
        first = False
    if valid_atis_airfare_departure_day != ["None"]:
        if not first: lm += ", "
        lm += "departure_day=" + capture(select(valid_atis_airfare_departure_day), name='dayName')
        consume_once("departure_day", lm['dayName'])
    return lm + ")"

@guidance(stateless=False)
def atis_flight_no(lm):
    first = True
    if valid_atis_flight_no_departure_city != ["None"]:
        lm += "departure_city=" + capture(select(valid_atis_flight_no_departure_city), name='fnDepName')
        consume_once("departure_city", lm['fnDepName'])
        first = False
    if valid_atis_flight_no_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_flight_no_destination_city), name='fnDestName')
        consume_once("destination_city", lm['fnDestName'])
        first = False
    if valid_atis_flight_no_airline != ["None"]:
        if not first: lm += ", "
        lm += "airline=" + capture(select(valid_atis_flight_no_airline), name='fnAirlineName')
        consume_once("airline", lm['fnAirlineName'])
        first = False
    if valid_atis_flight_no_relative_departure_time != ["None"]:
        if not first: lm += ", "
        lm += "relative_departure_time=" + capture(select(valid_atis_flight_no_relative_departure_time), name='fnRelTimeName')
        consume_once("relative_departure_time", lm['fnRelTimeName'])
        first = False
    if valid_atis_flight_no_departure_time != ["None"]:
        if not first: lm += ", "
        lm += "departure_time=" + capture(select(valid_atis_flight_no_departure_time), name='fnTimeName')
        consume_once("departure_time", lm['fnTimeName'])
        first = False
    if valid_atis_flight_no_departure_day != ["None"]:
        if not first: lm += ", "
        lm += "departure_day=" + capture(select(valid_atis_flight_no_departure_day), name='fnDayName')
        consume_once("departure_day", lm['fnDayName'])
    return lm + ")"

@guidance(stateless=False)
def atis_capacity(lm):
    first = True
    if valid_atis_capacity_airline != ["None"]:
        lm += "airline=" + capture(select(valid_atis_capacity_airline), name='capAirlineName')
        consume_once("airline", lm['capAirlineName'])
        first = False
    if valid_atis_capacity_departure_city != ["None"]:
        if not first: lm += ", "
        lm += "departure_city=" + capture(select(valid_atis_capacity_departure_city), name='capDepName')
        consume_once("departure_city", lm['capDepName'])
        first = False
    if valid_atis_capacity_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_capacity_destination_city), name='capDestName')
        consume_once("destination_city", lm['capDestName'])
        first = False
    if valid_atis_capacity_aircraft != ["None"]:
        if not first: lm += ", "
        lm += "aircraft=" + capture(select(valid_atis_capacity_aircraft), name='capAircraftName')
        consume_once("aircraft", lm['capAircraftName'])
    return lm + ")"

@guidance(stateless=False)
def atis_distance(lm):
    first = True
    if valid_atis_distance_departure_city != ["None"]:
        lm += "departure_city=" + capture(select(valid_atis_distance_departure_city), name='distDepName')
        consume_once("departure_city", lm['distDepName'])
        first = False
    if valid_atis_distance_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_distance_destination_city), name='distDestName')
        consume_once("destination_city", lm['distDestName'])
    return lm + ")"

@guidance(stateless=False)
def atis_ground_service(lm):
    first = True
    if valid_atis_ground_service_transport_type != ["None"]:
        lm += "transport_type=" + capture(select(valid_atis_ground_service_transport_type), name='transportName')
        consume_once("transport_type", lm['transportName'])
        first = False
    if valid_atis_ground_service_city != ["None"]:
        if not first: lm += ", "
        lm += "city=" + capture(select(valid_atis_ground_service_city), name='groundCityName')
        consume_once("city", lm['groundCityName'])
    return lm + ")"

@guidance(stateless=False)
def atis_abbreviation(lm):
    first = True
    if valid_atis_abbreviation_fare_code != ["None"]:
        lm += "fare_code=" + capture(select(valid_atis_abbreviation_fare_code), name='abbrFare')
        consume_once("fare_code", lm['abbrFare'])
        first = False
    if valid_atis_abbreviation_airline_code != ["None"]:
        if not first: lm += ", "
        lm += "airline_code=" + capture(select(valid_atis_abbreviation_airline_code), name='abbrAirline')
        consume_once("airline_code", lm['abbrAirline'])
        first = False
    if valid_atis_abbreviation_aircraft != ["None"]:
        if not first: lm += ", "
        lm += "aircraft=" + capture(select(valid_atis_abbreviation_aircraft), name='abbrCraft')
        consume_once("aircraft", lm['abbrCraft'])
    return lm + ")"

@guidance(stateless=False)
def atis_meal(lm):
    first = True
    if valid_atis_meal_flight_number != ["None"]:
        lm += "flight_number=" + capture(select(valid_atis_meal_flight_number), name='mealFlight')
        consume_once("flight_number", lm['mealFlight'])
        first = False
    if valid_atis_meal_meal_type != ["None"]:
        if not first: lm += ", "
        lm += "meal_type=" + capture(select(valid_atis_meal_meal_type), name='mealType')
        consume_once("meal_type", lm['mealType'])
        first = False
    if valid_atis_meal_departure_day != ["None"]:
        if not first: lm += ", "
        lm += "departure_day=" + capture(select(valid_atis_meal_departure_day), name='mealDay')
        consume_once("departure_day", lm['mealDay'])
        first = False
    if valid_atis_meal_departure_period != ["None"]:
        if not first: lm += ", "
        lm += "departure_period=" + capture(select(valid_atis_meal_departure_period), name='mealPeriod')
        consume_once("departure_period", lm['mealPeriod'])
        first = False
    if valid_atis_meal_meal_description != ["None"]:
        if not first: lm += ", "
        lm += "meal_description=" + capture(select(valid_atis_meal_meal_description), name='mealDesc')
        consume_once("meal_description", lm['mealDesc'])
        first = False
    if valid_atis_meal_airline != ["None"]:
        if not first: lm += ", "
        lm += "airline=" + capture(select(valid_atis_meal_airline), name='mealAirline')
        consume_once("airline", lm['mealAirline'])
    return lm + ")"

@guidance(stateless=False)
def atis_flight(lm):
    first = True
    if valid_atis_flight_departure_city != ["None"]:
        lm += "departure_city=" + capture(select(valid_atis_flight_departure_city), name='fltDepCity')
        consume_once("departure_city", lm['fltDepCity'])
        first = False
    if valid_atis_flight_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_flight_destination_city), name='fltDestCity')
        consume_once("destination_city", lm['fltDestCity'])
        first = False
    if valid_atis_flight_departure_month != ["None"]:
        if not first: lm += ", "
        lm += "departure_month=" + capture(select(valid_atis_flight_departure_month), name='fltMonth')
        consume_once("departure_month", lm['fltMonth'])
        first = False
    if valid_atis_flight_departure_day_number != ["None"]:
        if not first: lm += ", "
        lm += "departure_day_number=" + capture(select(valid_atis_flight_departure_day_number), name='fltDayNum')
        consume_once("departure_day_number", lm['fltDayNum'])
        first = False
    if valid_atis_flight_airline != ["None"]:
        if not first: lm += ", "
        lm += "airline=" + capture(select(valid_atis_flight_airline), name='fltAirline')
        consume_once("airline", lm['fltAirline'])
    return lm + ")"

@guidance(stateless=False)
def atis_flight_time(lm):
    first = True
    if valid_atis_flight_time_departure_city != ["None"]:
        lm += "departure_city=" + capture(select(valid_atis_flight_time_departure_city), name='ftDepCity')
        consume_once("departure_city", lm['ftDepCity'])
        first = False
    if valid_atis_flight_time_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_flight_time_destination_city), name='ftDestCity')
        consume_once("destination_city", lm['ftDestCity'])
        first = False
    if valid_atis_flight_time_flight_time != ["None"]:
        if not first: lm += ", "
        lm += "flight_time=" + capture(select(valid_atis_flight_time_flight_time), name='ftTime')
        consume_once("flight_time", lm['ftTime'])
    return lm + ")"

@guidance(stateless=False)
def atis_ground_fare(lm):
    first = True
    if valid_atis_ground_fare_transport_type != ["None"]:
        lm += "transport_type=" + capture(select(valid_atis_ground_fare_transport_type), name='gfTransport')
        consume_once("transport_type", lm['gfTransport'])
        first = False
    if valid_atis_ground_fare_city != ["None"]:
        if not first: lm += ", "
        lm += "city=" + capture(select(valid_atis_ground_fare_city), name='gfCity')
        consume_once("city", lm['gfCity'])
    return lm + ")"

@guidance(stateless=False)
def atis_airline(lm):
    first = True
    if valid_atis_airline_airline_code != ["None"]:
        lm += "airline_code=" + capture(select(valid_atis_airline_airline_code), name='alCode')
        consume_once("airline_code", lm['alCode'])
        first = False
    if valid_atis_airline_departure_city != ["None"]:
        if not first: lm += ", "
        lm += "departure_city=" + capture(select(valid_atis_airline_departure_city), name='alDepCity')
        consume_once("departure_city", lm['alDepCity'])
        first = False
    if valid_atis_airline_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_airline_destination_city), name='alDestCity')
        consume_once("destination_city", lm['alDestCity'])
    return lm + ")"

@guidance(stateless=False)
def atis_day_name(lm):
    first = True
    if valid_atis_day_name_departure_city != ["None"]:
        lm += "departure_city=" + capture(select(valid_atis_day_name_departure_city), name='dnDepCity')
        consume_once("departure_city", lm['dnDepCity'])
        first = False
    if valid_atis_day_name_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_day_name_destination_city), name='dnDestCity')
        consume_once("destination_city", lm['dnDestCity'])
    return lm + ")"

@guidance(stateless=False)
def atis_aircraft(lm):
    first = True
    if valid_atis_aircraft_aircraft != ["None"]:
        lm += "aircraft=" + capture(select(valid_atis_aircraft_aircraft), name='craftName')
        consume_once("aircraft", lm['craftName'])
        first = False
    if valid_atis_aircraft_departure_city != ["None"]:
        if not first: lm += ", "
        lm += "departure_city=" + capture(select(valid_atis_aircraft_departure_city), name='acDepCity')
        consume_once("departure_city", lm['acDepCity'])
        first = False
    if valid_atis_aircraft_destination_city != ["None"]:
        if not first: lm += ", "
        lm += "destination_city=" + capture(select(valid_atis_aircraft_destination_city), name='acDestCity')
        consume_once("destination_city", lm['acDestCity'])
        first = False
    if valid_atis_aircraft_relative_departure_time != ["None"]:
        if not first: lm += ", "
        lm += "relative_departure_time=" + capture(select(valid_atis_aircraft_relative_departure_time), name='acRelTime')
        consume_once("relative_departure_time", lm['acRelTime'])
        first = False
    if valid_atis_aircraft_departure_period != ["None"]:
        if not first: lm += ", "
        lm += "departure_period=" + capture(select(valid_atis_aircraft_departure_period), name='acDepPeriod')
        consume_once("departure_period", lm['acDepPeriod'])
        first = False
    if valid_atis_aircraft_airline != ["None"]:
        if not first: lm += ", "
        lm += "airline=" + capture(select(valid_atis_aircraft_airline), name='acAirline')
        consume_once("airline", lm['acAirline'])
        first = False
    if valid_atis_aircraft_departure_time != ["None"]:
        if not first: lm += ", "
        lm += "departure_time=" + capture(select(valid_atis_aircraft_departure_time), name='acDepTime')
        consume_once("departure_time", lm['acDepTime'])
    return lm + ")"


# === the list‐builder at the end ===
@guidance(stateless=False)
def valid_atis_call(lm):
    lm += "["
    funcs = [
                "atis_airport",
                "atis_city",
                "atis_quantity",
                "atis_airfare",
                "atis_flight_no",
                "atis_capacity",
                "atis_distance",
                "atis_ground_service",
                "atis_abbreviation",
                "atis_meal",
                "atis_flight",
                "atis_flight_time",
                "atis_ground_fare",
                "atis_airline",
                "atis_day_name",
                "atis_aircraft"]

    for i in range(3):

        lm += select(funcs, name=f"choice_{i}")
        choice = lm[f"choice_{i}"]
        if choice == "]":
            return lm
        lm += "("

        if lm[f"choice_{i}"] == "]":
            return lm

        if   lm[f"choice_{i}"] == "atis_airport":
            lm += atis_airport()
        elif lm[f"choice_{i}"] == "atis_city":
            lm += atis_city()
        elif lm[f"choice_{i}"] == "atis_quantity":
            lm += atis_quantity()
        elif lm[f"choice_{i}"] == "atis_airfare":
            lm += atis_airfare()
        elif lm[f"choice_{i}"] == "atis_flight_no":
            lm += atis_flight_no()
        elif lm[f"choice_{i}"] == "atis_capacity":
            lm += atis_capacity()
        elif lm[f"choice_{i}"] == "atis_distance":
            lm += atis_distance()
        elif lm[f"choice_{i}"] == "atis_ground_service":
            lm += atis_ground_service()
        elif lm[f"choice_{i}"] == "atis_abbreviation":
            lm += atis_abbreviation()
        elif lm[f"choice_{i}"] == "atis_meal":
            lm += atis_meal()
        elif lm[f"choice_{i}"] == "atis_flight":
            lm += atis_flight()
        elif lm[f"choice_{i}"] == "atis_flight_time":
            lm += atis_flight_time()
        elif lm[f"choice_{i}"] == "atis_ground_fare":
            lm += atis_ground_fare()
        elif lm[f"choice_{i}"] == "atis_airline":
            lm += atis_airline()
        elif lm[f"choice_{i}"] == "atis_day_name":
            lm += atis_day_name()
        elif lm[f"choice_{i}"] == "atis_aircraft":
            lm += atis_aircraft()

        lm += select([", ", "]"], name=f"sep_{i}")
        # lm += select([", ", "]"], name=f"sep_{i}")
        if lm[f"sep_{i}"] == "]":
            return lm
    lm += "]"
    return lm
```
