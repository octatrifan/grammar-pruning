```python
import guidance
from guidance import select, capture, optional

for intent, slots in schema.items():
    for slot, values in slots.items():
        list_name = f"valid_{intent}_{slot}"
        globals()[list_name] = [f"'{val}'" for val in values] + ["None"]

for intent, slots in schema.items():
    for slot in slots:
        list_name = f"valid_{intent}_{slot}"
        print(f"{list_name} = {globals()[list_name]}")

def consume_once(param, val):
    """
    Remove a used slot value from every valid_<Intent>_<slot> list
    in the global namespace, so it won’t be picked again anywhere.
    """
    if val == "None":
        return

    for name, lst in globals().items():
        # target only our valid_* lists
        if name.startswith("valid_") and isinstance(lst, list):
            # if the value is present, remove it
            if val in lst:
                lst.remove(val)

@guidance(stateless=False)
def AddToPlaylist(lm):
    first = True
    if valid_AddToPlaylist_playlist != ["None"]:
        lm += "playlist=" + capture(select(valid_AddToPlaylist_playlist), name="playlistVal")
        consume_once("playlist", lm["playlistVal"])
        first = False
    if valid_AddToPlaylist_playlist_owner != ["None"]:
        if not first: lm += ", "
        lm += "playlist_owner=" + capture(select(valid_AddToPlaylist_playlist_owner), name="playlistOwnerVal")
        consume_once("playlist_owner", lm["playlistOwnerVal"])
        first = False
    if valid_AddToPlaylist_entity_name != ["None"]:
        if not first: lm += ", "
        lm += "entity_name=" + capture(select(valid_AddToPlaylist_entity_name), name="entityNameVal")
        consume_once("entity_name", lm["entityNameVal"])
        first = False
    if valid_AddToPlaylist_artist != ["None"]:
        if not first: lm += ", "
        lm += "artist=" + capture(select(valid_AddToPlaylist_artist), name="artistVal")
        consume_once("artist", lm["artistVal"])
        first = False
    if valid_AddToPlaylist_music_item != ["None"]:
        if not first: lm += ", "
        lm += "music_item=" + capture(select(valid_AddToPlaylist_music_item), name="musicItemVal")
        consume_once("music_item", lm["musicItemVal"])
    return lm + ")"


@guidance(stateless=False)
def PlayMusic(lm):
    first = True
    if valid_PlayMusic_artist != ["None"]:
        lm += "artist=" + capture(select(valid_PlayMusic_artist), name="artistVal")
        consume_once("artist", lm["artistVal"])
        first = False
    if valid_PlayMusic_sort != ["None"]:
        if not first: lm += ", "
        lm += "sort=" + capture(select(valid_PlayMusic_sort), name="sortVal")
        consume_once("sort", lm["sortVal"])
        first = False
    if valid_PlayMusic_year != ["None"]:
        if not first: lm += ", "
        lm += "year=" + capture(select(valid_PlayMusic_year), name="yearVal")
        consume_once("year", lm["yearVal"])
        first = False
    if valid_PlayMusic_music_item != ["None"]:
        if not first: lm += ", "
        lm += "music_item=" + capture(select(valid_PlayMusic_music_item), name="musicItemVal")
        consume_once("music_item", lm["musicItemVal"])
        first = False
    if valid_PlayMusic_playlist != ["None"]:
        if not first: lm += ", "
        lm += "playlist=" + capture(select(valid_PlayMusic_playlist), name="playlistVal")
        consume_once("playlist", lm["playlistVal"])
        first = False
    if valid_PlayMusic_service != ["None"]:
        if not first: lm += ", "
        lm += "service=" + capture(select(valid_PlayMusic_service), name="serviceVal")
        consume_once("service", lm["serviceVal"])
        first = False
    if valid_PlayMusic_genre != ["None"]:
        if not first: lm += ", "
        lm += "genre=" + capture(select(valid_PlayMusic_genre), name="genreVal")
        consume_once("genre", lm["genreVal"])
        first = False
    if valid_PlayMusic_album != ["None"]:
        if not first: lm += ", "
        lm += "album=" + capture(select(valid_PlayMusic_album), name="albumVal")
        consume_once("album", lm["albumVal"])
    return lm + ")"


@guidance(stateless=False)
def BookRestaurant(lm):
    first = True
    if valid_BookRestaurant_spatial_relation != ["None"]:
        lm += "spatial_relation=" + capture(select(valid_BookRestaurant_spatial_relation), name="spatialRelationVal")
        consume_once("spatial_relation", lm["spatialRelationVal"])
        first = False
    if valid_BookRestaurant_city != ["None"]:
        if not first: lm += ", "
        lm += "city=" + capture(select(valid_BookRestaurant_city), name="cityVal")
        consume_once("city", lm["cityVal"])
        first = False
    if valid_BookRestaurant_time != ["None"]:
        if not first: lm += ", "
        lm += "time=" + capture(select(valid_BookRestaurant_time), name="timeVal")
        consume_once("time", lm["timeVal"])
        first = False
    if valid_BookRestaurant_party_size_number != ["None"]:
        if not first: lm += ", "
        lm += "party_size_number=" + capture(select(valid_BookRestaurant_party_size_number), name="partySizeNumberVal")
        consume_once("party_size_number", lm["partySizeNumberVal"])
        first = False
    if valid_BookRestaurant_restaurant_type != ["None"]:
        if not first: lm += ", "
        lm += "restaurant_type=" + capture(select(valid_BookRestaurant_restaurant_type), name="restaurantTypeVal")
        consume_once("restaurant_type", lm["restaurantTypeVal"])
        first = False
    if valid_BookRestaurant_served_dish != ["None"]:
        if not first: lm += ", "
        lm += "served_dish=" + capture(select(valid_BookRestaurant_served_dish), name="servedDishVal")
        consume_once("served_dish", lm["servedDishVal"])
        first = False
    if valid_BookRestaurant_country != ["None"]:
        if not first: lm += ", "
        lm += "country=" + capture(select(valid_BookRestaurant_country), name="countryVal")
        consume_once("country", lm["countryVal"])
        first = False
    if valid_BookRestaurant_restaurant_name != ["None"]:
        if not first: lm += ", "
        lm += "restaurant_name=" + capture(select(valid_BookRestaurant_restaurant_name), name="restaurantNameVal")
        consume_once("restaurant_name", lm["restaurantNameVal"])
        first = False
    if valid_BookRestaurant_timeRange != ["None"]:
        if not first: lm += ", "
        lm += "timeRange=" + capture(select(valid_BookRestaurant_timeRange), name="timeRangeVal")
        consume_once("timeRange", lm["timeRangeVal"])
        first = False
    if valid_BookRestaurant_state != ["None"]:
        if not first: lm += ", "
        lm += "state=" + capture(select(valid_BookRestaurant_state), name="stateVal")
        consume_once("state", lm["stateVal"])
        first = False
    if valid_BookRestaurant_cuisine != ["None"]:
        if not first: lm += ", "
        lm += "cuisine=" + capture(select(valid_BookRestaurant_cuisine), name="cuisineVal")
        consume_once("cuisine", lm["cuisineVal"])
    return lm + ")"


@guidance(stateless=False)
def SearchCreativeWork(lm):
    first = True
    if valid_SearchCreativeWork_object_type != ["None"]:
        lm += "object_type=" + capture(select(valid_SearchCreativeWork_object_type), name="objectTypeVal")
        consume_once("object_type", lm["objectTypeVal"])
        first = False
    if valid_SearchCreativeWork_object_name != ["None"]:
        if not first: lm += ", "
        lm += "object_name=" + capture(select(valid_SearchCreativeWork_object_name), name="objectNameVal")
        consume_once("object_name", lm["objectNameVal"])
    return lm + ")"


@guidance(stateless=False)
def SearchScreeningEvent(lm):
    first = True
    if valid_SearchScreeningEvent_movie_name != ["None"]:
        lm += "movie_name=" + capture(select(valid_SearchScreeningEvent_movie_name), name="movieNameVal")
        consume_once("movie_name", lm["movieNameVal"])
        first = False
    if valid_SearchScreeningEvent_location_name != ["None"]:
        if not first: lm += ", "
        lm += "location_name=" + capture(select(valid_SearchScreeningEvent_location_name), name="locationNameVal")
        consume_once("location_name", lm["locationNameVal"])
        first = False
    if valid_SearchScreeningEvent_movie_type != ["None"]:
        if not first: lm += ", "
        lm += "movie_type=" + capture(select(valid_SearchScreeningEvent_movie_type), name="movieTypeVal")
        consume_once("movie_type", lm["movieTypeVal"])
        first = False
    if valid_SearchScreeningEvent_spatial_relation != ["None"]:
        if not first: lm += ", "
        lm += "spatial_relation=" + capture(select(valid_SearchScreeningEvent_spatial_relation), name="spatialRelationVal")
        consume_once("spatial_relation", lm["spatialRelationVal"])
        first = False
    if valid_SearchScreeningEvent_timeRange != ["None"]:
        if not first: lm += ", "
        lm += "timeRange=" + capture(select(valid_SearchScreeningEvent_timeRange), name="timeRangeVal")
        consume_once("timeRange", lm["timeRangeVal"])
    return lm + ")"


@guidance(stateless=False)
def RateBook(lm):
    first = True
    if valid_RateBook_object_name != ["None"]:
        lm += "object_name=" + capture(select(valid_RateBook_object_name), name="objectNameVal")
        consume_once("object_name", lm["objectNameVal"])
        first = False
    if valid_RateBook_rating_value != ["None"]:
        if not first: lm += ", "
        lm += "rating_value=" + capture(select(valid_RateBook_rating_value), name="ratingValueVal")
        consume_once("rating_value", lm["ratingValueVal"])
        first = False
    if valid_RateBook_rating_unit != ["None"]:
        if not first: lm += ", "
        lm += "rating_unit=" + capture(select(valid_RateBook_rating_unit), name="ratingUnitVal")
        consume_once("rating_unit", lm["ratingUnitVal"])
        first = False
    if valid_RateBook_best_rating != ["None"]:
        if not first: lm += ", "
        lm += "best_rating=" + capture(select(valid_RateBook_best_rating), name="bestRatingVal")
        consume_once("best_rating", lm["bestRatingVal"])
        first = False
    if valid_RateBook_object_type != ["None"]:
        if not first: lm += ", "
        lm += "object_type=" + capture(select(valid_RateBook_object_type), name="objectTypeVal")
        consume_once("object_type", lm["objectTypeVal"])
        first = False
    if valid_RateBook_object_part_of_series_type != ["None"]:
        if not first: lm += ", "
        lm += "object_part_of_series_type=" + capture(select(valid_RateBook_object_part_of_series_type), name="objectPartOfSeriesTypeVal")
        consume_once("object_part_of_series_type", lm["objectPartOfSeriesTypeVal"])
    return lm + ")"


@guidance(stateless=False)
def GetWeather(lm):
    first = True
    if valid_GetWeather_city != ["None"]:
        lm += "city=" + capture(select(valid_GetWeather_city), name="cityVal")
        consume_once("city", lm["cityVal"])
        first = False
    if valid_GetWeather_geographic_poi != ["None"]:
        if not first: lm += ", "
        lm += "geographic_poi=" + capture(select(valid_GetWeather_geographic_poi), name="geographicPoiVal")
        consume_once("geographic_poi", lm["geographicPoiVal"])
        first = False
    if valid_GetWeather_timeRange != ["None"]:
        if not first: lm += ", "
        lm += "timeRange=" + capture(select(valid_GetWeather_timeRange), name="timeRangeVal")
        consume_once("timeRange", lm["timeRangeVal"])
        first = False
    if valid_GetWeather_condition_description != ["None"]:
        if not first: lm += ", "
        lm += "condition_description=" + capture(select(valid_GetWeather_condition_description), name="conditionDescriptionVal")
        consume_once("condition_description", lm["conditionDescriptionVal"])
        first = False
    if valid_GetWeather_state != ["None"]:
        if not first: lm += ", "
        lm += "state=" + capture(select(valid_GetWeather_state), name="stateVal")
        consume_once("state", lm["stateVal"])
        first = False
    if valid_GetWeather_country != ["None"]:
        if not first: lm += ", "
        lm += "country=" + capture(select(valid_GetWeather_country), name="countryVal")
        consume_once("country", lm["countryVal"])
        first = False
    if valid_GetWeather_condition_temperature != ["None"]:
        if not first: lm += ", "
        lm += "condition_temperature=" + capture(select(valid_GetWeather_condition_temperature), name="conditionTemperatureVal")
        consume_once("condition_temperature", lm["conditionTemperatureVal"])
    return lm + ")"

@guidance(stateless=False)
def valid_mixsnips_call(lm, max_intents=3):
    """
    Build a list of up to `max_intents` MixSNIPS intent calls,
    selecting each only while it has remaining slot values.
    """
    lm += "["
    intents = [
        "AddToPlaylist",
        "PlayMusic",
        "BookRestaurant",
        "SearchCreativeWork",
        "SearchScreeningEvent",
        "RateBook",
        "GetWeather"
    ]
    for i in range(max_intents):
        # stop if every valid_* list has been exhausted
        if all(
            globals()[name] == ["None"]
            for name in globals()
            if name.startswith("valid_")
        ):
            lm += "]"
            return lm

        # collect intents which still have at least one non-None slot
        available = []
        for intent in intents:
            if intent in ("SearchCreativeEvent"):
                # for these two, require an object_name
                key = f"valid_{intent}_object_name"
                if globals().get(key, ["None"]) != ["None"]:
                    available.append(f"{intent}(")
            else:
                # any other intent can fire if any of its slots is non‐None
                if any(
                    globals()[f"valid_{intent}_{slot}"] != ["None"]
                    for slot in schema[intent]
                ):
                    available.append(f"{intent}(")

        # choose the next intent to emit
        lm += select(available, name=f"choice_{i}")
        choice = lm[f"choice_{i}"]
        if choice == "]":
            lm += "]"
            return lm

        # invoke the corresponding intent function
        if choice == "AddToPlaylist(":
            lm += AddToPlaylist()
        elif choice == "PlayMusic(":
            lm += PlayMusic()
        elif choice == "BookRestaurant(":
            lm += BookRestaurant()
        elif choice == "SearchCreativeWork(":
            lm += SearchCreativeWork()
        elif choice == "SearchScreeningEvent(":
            lm += SearchScreeningEvent()
        elif choice == "RateBook(":
            lm += RateBook()
        elif choice == "GetWeather(":
            lm += GetWeather()

        lm += ", "
        # lm += select([", ", "]"], name=f"sep_{i}")
        # if lm[f"sep_{i}"] == "]":
        #     break

    lm += "]"
    return lm
```

```python
import re
import json

param_values = {}
for func_params in schema.values():
    for param, vals in func_params.items():
        param_values.setdefault(param, set()).update(vals)

def extract_input_items(text):
    """
    Allows the same value to be matched under multiple parameters.
    For MixSNIPS and MixATIS this is a very simple extraction
    Note: We can use more advanced techniques (regex for date/time, numbers, etc)
    However, this is very dependent on the use-case.
    """
    found = {}
    lc = text.lower()
    matched_ranges_by_param = {param: set() for param in param_values}  # Track ranges per parameter

    all_values = sorted(
        [(param, v) for param, vals in param_values.items() for v in vals],
        key=lambda x: len(x[1]),
        reverse=True
    )

    for param, v in all_values:
        for match in re.finditer(r'\b' + re.escape(v.lower()) + r'\b', lc):
            start, end = match.start(), match.end()
            if any(existing_start <= start and existing_end >= end for existing_start, existing_end in matched_ranges_by_param[param]):
                continue
            matched_ranges_by_param[param].add((start, end))
            found.setdefault(param, []).append(v)

    return found

def extract_output_items(calls):
    results = []
    pat = re.compile(r"(\w+)\(([^)]*)\)")
    for call in calls:
        m = pat.match(call)
        if not m:
            continue
        func, body = m.groups()
        params = {}
        for part in re.split(r",\s*", body):
            if "=" in part:
                k, v = part.split("=", 1)
                params[k.strip()] = v.strip().strip("'")
        results.append({func: params})
    return results

def augment_data(in_path="snips_data.json", out_path="snips_data_augmented.json"):
    with open(in_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    with open(out_path, 'w', encoding='utf-8') as out:
        for entry in data:
            if not isinstance(entry, dict):
                continue
            entry["input_items"] = extract_input_items(entry.get("input", ""))
            entry["output_items"] = extract_output_items(entry.get("output", []))
            out.write(json.dumps(entry) + "\n")
    print(f"Augmented data written to {out_path}")


augment_data()
```
